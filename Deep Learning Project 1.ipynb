```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
tf.__version__
```

```python
mnist = tf.keras.datasets.mnist # 28x28 data of images with hand written digits 0-9
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print(x_train[0])
```

```python
plt.imshow(x_train[0])
plt.show
```

```python
# Normalize the data or Scale down the data

x_train = tf.keras.utils.normalize(x_train)
```

```python
model = tf.keras.models.Sequential() # model
model.add(tf.keras.layers.Flatten()) # Add Input Layers
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu)) # Add Hidden - 1 Layers
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu)) # Add Hidden - 2 Layers
model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax)) # Add Output Layers

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=3)
```

```python
val_loss, val_acc = model.evaluate(x_test, y_test)
print(val_loss, val_acc)
```

```python
prediction = model.predict([x_test])
```
